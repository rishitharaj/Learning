**3. Create a numpy array containing the numbers from 1 to 20, and then extract the elements between the 5th and 15th index.**

```python
arr_1 = np.arange(1,21)
print(arr_1)
print(arr_1[5:16])
```

**4. How to compute the mean, median, standard deviation of a numpy array?Use the array you created above**

```python
print("Mean:",np.mean(arr_1))
print("Median:",np.median(arr_1))
print("Standard Deviation:",np.std(arr_1))
```
